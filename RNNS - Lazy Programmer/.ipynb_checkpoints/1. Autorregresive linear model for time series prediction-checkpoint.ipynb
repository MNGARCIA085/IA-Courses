## <center> <font color='blue'> <b> Autorregresive linear model for time series prediction </b> </font> </center>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as mae

from keras.models import Model
from keras.layers import Input, LSTM, GRU, SimpleRNN, Dense
import keras.backend as K

from keras.optimizers import SGD, Adam
```

```python
# make the data, a sine wave with noise
series = np.sin(0.1 * np.arange(200)) + np.random.rand(200)*0.1
```

```python
# plot it
plt.plot(series)
plt.show()
```

```python
### build the dataset
# let's see if we can use T past values to predict the next value

T = 10
D = 1
X = []
Y = []

for t in range(len(series) - T):
    x = series[t:t+T]
    X.append(x)
    y = series[t+T]
    Y.append(y)

X = np.array(X) # cast to numpy array
Y = np.array(Y)
N = len(X)
    
```

```python
X.shape, Y.shape
```

```python
# split into train and test
# because is a forecasting problem we dont split randomly

X_train, y_train = X[:-N//2], Y[:-N//2]
X_val, y_val = X[-N//2:], Y[-N//2:]
```

```python
# model
i = Input(shape=(T, ))
x = Dense(1)(i)
model = Model(i, x)

model.compile(
        loss = 'mse',
        optimizer = Adam(learning_rate=0.1)
)

history = model.fit(
        X_train, y_train,
        epochs = 80,
        verbose = 0,
        validation_data = (X_val, y_val)
)
```

```python
# Access the history attributes
history_dict = history.history

# List available metrics
print(history_dict.keys())
```

```python
history_dict['loss'][-1], history_dict['val_loss'][-1]
```

```python
# Extract loss and accuracy from the history dictionary
loss = history_dict['loss']
val_loss = history_dict['val_loss']

# If you have accuracy as a metric (for classification)
accuracy = history_dict.get('accuracy')
val_accuracy = history_dict.get('val_accuracy')

# Create a range for the number of epochs
epochs = range(1, len(loss) + 1)

# Plot training and validation loss
plt.figure(figsize=(12, 4))

# Loss plot
# Loss plot
plt.subplot(1, 2, 1)
plt.plot(epochs, loss, color='red', label='Training loss')  # Change to green
plt.plot(epochs, val_loss, color='orange', label='Validation loss')  # Change to red
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()


# Accuracy plot (if available)
if accuracy:
    plt.subplot(1, 2, 2)
    plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
    plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
    plt.title('Training and validation accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

plt.show()
```

### <b> <font color='red'> Incorrect way to forecast </font> </b>

```python
val_pred = []

#index of first validation input
i = 0

while (len(val_pred) < len(y_val)):
    p = model.predict(X_val[i].reshape(1, -1), verbose=0)[0,0] # 1x1 array -> scalar
    i += 1
    
    # update predict list
    val_pred.append(p)
```

```python
plt.plot(y_val, label = 'forecast_target')
plt.plot(val_pred, label = 'forecast prediction')
plt.legend()
```

### <b> <font color='blue'> Correct way to forecast </font> </b>

```python
# to predict yhat1 I use as input xval1 (which I suppose I know)
# to predict yhat2 I use, as input yhat1 (the prediction obtained with xval1)
# to predict yhat3 I use as input yhat2 ....


val_pred_ok = []

last_x = X_val[0]


while ( len(val_pred_ok) < len(y_val)) :
    p = model.predict(last_x.reshape(1, -1), verbose=0)[0,0]
    val_pred_ok.append(p)

    # make the new input
    last_x = np.roll(last_x, -1)
    last_x[-1] = p
```

```python
plt.plot(y_val, label = 'forecast_target')
plt.plot(val_pred_ok, label = 'forecast prediction')
plt.legend()
```

### Recursive forecasting

In the recursive forecasting (or auto-regressive forecasting), you use the previously predicted values as inputs for future predictions. This method is common when you're forecasting multiple time steps into the future without knowing the true future values.

To clarify:
- To predict yhat1, you use xval1, which is the real known data.
- To predict yhat2, you use yhat1, which is the prediction obtained using xval1.
- To predict yhat3, you use yhat2, and so on.

This is a common method in time series forecasting when using models like Recurrent Neural Networks (RNNs), LSTMs, or even simpler linear models, where each step depends on the previous prediction.

A potential issue:
Error propagation: In recursive forecasting, each prediction relies on the previous prediction. This can cause errors to accumulate as you move further into the forecast, because if one prediction is slightly off, it can negatively impact subsequent predictions.

Keep in mind that the further you forecast, the greater the risk of error accumulation, which can make predictions less accurate over time.
